```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define directories
INPUT_DIR = "input"
OUTPUT_DIR = "output/01-preprocessing"

# Create directories if they don't exist
os.makedirs(INPUT_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
image_filename = "0008.png"
input_path = os.path.join(INPUT_DIR, image_filename)

# Read the original image
original_img = cv2.imread(input_path)
```

```python
# Helper function to display images
def show_image(image, title="Image", cmap_type="gray"):
    plt.figure(figsize=(6, 4))
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# -------------------- Grayscale --------------------
# Convert to Grayscale
gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

# Save grayscale image
gray_output_path = os.path.join(OUTPUT_DIR, "01-grayscale-" + image_filename)
cv2.imwrite(gray_output_path, gray_img)
print(f"Grayscale image saved to: {gray_output_path}")

# 4. Display the result in the notebook
show_image(gray_img, title="Step 1: Grayscale Conversion")
```

```python
# -------------------- Binarisation --------------------

# Adaptive Gaussian Thresholding
adaptive_img = cv2.adaptiveThreshold(
    gray_img,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    21,
    2,
)

# Save binarised image
adaptive_path = os.path.join(OUTPUT_DIR, "02-binarised-" + image_filename)
cv2.imwrite(adaptive_path, adaptive_img)
print(f"Binarised image saved to: {adaptive_path}")

# Display results
show_image(adaptive_img, title="Step 2: Adaptive Gaussian Threshold")
```

```python
# -------------------- Padding --------------------

# Padding size
PAD = 4

# Add white border
padded = cv2.copyMakeBorder(
    adaptive_img,
    top=PAD,
    bottom=PAD,
    left=PAD,
    right=PAD,
    borderType=cv2.BORDER_CONSTANT,
    value=255,
)

# Save output
padded_path = os.path.join(OUTPUT_DIR, "03-padded-" + image_filename)
cv2.imwrite(padded_path, padded)
print(f"Padded image saved to: {padded_path}")

# Display
show_image(padded, title="Step 3: Padded Image")
```

```python
# -------------------- Finalise Output Image --------------------

final_path = os.path.join(OUTPUT_DIR, "preprocessed-" + image_filename)
cv2.imwrite(final_path, padded)
print(f"Final image saved to: {final_path}")
```
