```python
import os
import pandas as pd
from tqdm import tqdm
import json

# Create folder
os.makedirs("dataset/images/test/images", exist_ok=True)
metadata = []
```

```python
# Read data
df = pd.read_parquet("dataset/data/test.parquet")
df = df.rename(columns={"image.bytes": "image_bytes", "image.path": "image_path"})
```

```python
# Save images
for row in tqdm(df.itertuples(index=False), total=len(df)):
    img_bytes = row.image_bytes
    text = row.label

    name, ext = os.path.splitext(row.filename)
    filename = f"{int(name):04d}{ext}"

    with open(f"dataset/images/test/images/{filename}", "wb") as f:
        f.write(img_bytes)

    metadata.append({"filename": filename, "text": text})

# Save metadata
with open("dataset/images/test/metadata.json", "w", encoding="utf-8") as f:
    json.dump(metadata, f, ensure_ascii=False, indent=2)
```
